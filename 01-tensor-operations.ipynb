## Function 1 - `torch.randn`

#### **Syntax** :-torch.randn(*size, out=None, dtype=None, layout=torch.strided, device=None, requires_grad=False) → Tensor



```python
t1 = torch.randn(5)
print(t1)
```

```python
x = torch.randn(3, 4, (5))
print(x)
```

*Returns a tensor filled with random numbers from a normal distribution
with mean `0` and variance `1` (also called the standard normal
distribution).*

## Function 2 - `randint`

tensor.randint(low=0, high, size, *, generator=None, out=None,dtype=None, layout=torch.strided, device=None, requires_grad=False)

**Returns a tensor filled with random integers generated uniformly between low (inclusive) and high (exclusive).**

```python
torch.randint(10, (2, 2))
```

Returns Random Numbers in Int.

```python
torch.randint(3, 10, (2, 2))
```

Run IT.

## Function 3 - torch.logical_and()
torch.logical_and(input, other, out=None) → Tensor

**Computes the element-wise logical AND of the given input tensors. Zeros are treated as False and nonzeros are treated as True.**

```python
torch.logical_and(torch.tensor([True, False, True]), torch.tensor([True, False, False]))
tensor([ True, False, False])
a = torch.tensor([0, 1, 10, 0], dtype=torch.int8)
b = torch.tensor([4, 0, 1, 0], dtype=torch.int8)
torch.logical_and(a, b)
```

Computes the element-wise logical AND of the given input tensors. Zeros are treated as False and nonzeros are treated as True.

```python
torch.logical_and(a.double(), b.double())
```

```python
torch.logical_and(a, b, out=torch.empty(4, dtype=torch.bool))
```

## Function 4 - `torch.mul()`

**Multiplies each element of the input input with the scalar other and returns a new resulting tensor.**

```python
torch.mul(5, 100)
```

    Multiplies the N

```python
a = torch.randn(3)
torch.mul(a, 100)
```

Just Run It.

## Function 5 - `torch.neg()`

**Returns a new tensor with the negative of the elements of input.**

```python
a = torch.randn(5)
torch.neg(a)
```

Return Random number & then converts +ve to -ve,vice-versa

```python
a = torch.randn(5)
torch.neg(a)
```

Return Negative into Positive,vice-versa
